Bag of Words + TF-IDF + Word2Vec

```python
# Assignment 2: BoW, TF-IDF, Word2Vec

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
```

```python
documents = [
    "I love machine learning",
    "Machine learning is powerful",
    "Natural language processing is part of AI"
]
```

```python
# Bag of Words (Count)
count_vec = CountVectorizer()
bow = count_vec.fit_transform(documents)
print("BoW Count:\n", bow.toarray())
```

```python
# Normalized BoW
count_vec = CountVectorizer()
bow = count_vec.fit_transform(documents)

print("BoW (Count):")
print(bow.toarray())
```

```python
# TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)
print("TF-IDF:\n", tfidf_matrix.toarray())
```

```python
# Word2Vec Embeddings
tokenized_docs = [doc.split() for doc in documents]
w2v_model = Word2Vec(tokenized_docs, vector_size=50, window=3, min_count=1)

print("Word2Vec for 'machine':\n", w2v_model.wv['machine'])
```
