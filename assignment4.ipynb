Named Entity Recognition (NER) + Evaluation Metrics


```python
# Assignment 4: NER using spaCy

import spacy
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
# Load model
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Apple is looking at buying U.K. startup for $1 billion."
```

```python
doc = nlp(text)
```

```python
# Extract entities
entities = [(ent.text, ent.label_) for ent in doc.ents]
print("Named Entities:", entities)
```

```python
# Ground Truth (example)
true_labels = ['ORG', 'GPE', 'MONEY']
predicted_labels = [ent.label_ for ent in doc.ents]
```

```python
# Convert to numeric for metrics
label_map = {label: i for i, label in enumerate(set(true_labels + predicted_labels))}
true_encoded = [label_map[l] for l in true_labels]
pred_encoded = [label_map[l] for l in predicted_labels]
```

```python
# Evaluation
print("Accuracy:", accuracy_score(true_encoded, pred_encoded))
print("Precision:", precision_score(true_encoded, pred_encoded, average='macro'))
print("Recall:", recall_score(true_encoded, pred_encoded, average='macro'))
print("F1 Score:", f1_score(true_encoded, pred_encoded, average='macro'))
```
